```python
common_symptoms = [
    "Fatigue",
    "Fever",
    "Nausea",
    "Shortness of breath",
    "Cough",
    "Headache",
    "Vomiting",
    "Joint pain",
    "Swelling",
    "Weakness"   
]
```

```python
symptoms=[]
asked=[]
considered=[]
flag=True
print("Hi, I am Dr. Hancock.")
print("I can diagnose what disease you may have from your symptoms, but I sugest you to consult a doctor as soon as possible")

while True:
    if flag:
        try:
            symptom= choice(list(set(common_symptoms)-set(asked)))
        
        except: 
            flag=False
                    
    else:
        symptom= choice(list(set(diseases[choice(considered)])-set(asked)))
           
    print()
    print("Do you have",symptom)
    ch=input('y/n >> ')
    asked.append(symptom)
    if ch in ('y','Y'): 
        symptoms.append(symptom)
        if symptom in common_symptoms: flag= False

    elif ch in ('n','N'): pass
    else:
        print('Invalid input! please enter "y" or "n"')
        continue
 
    for i in diseases:
        if set(diseases[i])&set(symptoms)==set(symptoms):
            if i not in considered: considered.append(i)
        else: 
            if i in considered: considered.remove(i)
                
    if len(considered)==1:
        print()
        symptom= choice(list(set(diseases[choice(considered)])-set(asked)))
        print("Do you have",symptom)
        if input('y/n >> ') in ('y','Y'):
            print('You might have',considered[0])
        else:
            print("Sorry, but I couldn't figure out what you may have. I guess you might have",considered[0], 
                  "but, I suggest you to contact a doctor")
        break
```
